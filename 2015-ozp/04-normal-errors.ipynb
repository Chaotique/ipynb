```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import preprocessing
import Orange
from scipy.stats import norm
%matplotlib inline
```

# Porazdelitev napak napovednega modela linearne regresije na učni množici

```python
housing = Orange.data.Table("housing")
```

```python
domain = Orange.data.Domain(housing.domain.attributes[4:5], housing.domain.class_vars)
data = Orange.data.Table(domain, housing)
```

```python
lr = Orange.regression.RidgeRegressionLearner()
model = lr(data)
```

```python
err = model(data) - data.Y
```

```python
plt.hist(err, 50, normed=True, color="yellow");
mu, std = norm.fit(err)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2);
```

Stranpot: normalna porazdelitev

```python
def normal(x, mu=0, sigma=1):
    return (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-(x-mu)**2/(2*sigma))
```

```python
x = np.linspace(-3, 3, 100)
plt.plot(x, normal(x));
```

```python
plt.plot(x, norm.pdf(x));
```
