```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = lambda a: (a-5)**2+3
a = np.arange(0,10,0.1)
y = f(a)
```

```python
plt.plot(a, y, "k-");
```

```python
def derivative(f, a, eps=1e-3):
    return (f(a+eps) - f(a-eps)) / (2*eps)
```

```python
ai = 8
for _ in range(10):
    ai = ai - 0.2 * derivative(f, ai)
    print("%.2f" % ai)
```
